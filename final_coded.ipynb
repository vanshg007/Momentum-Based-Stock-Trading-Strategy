```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from ta.momentum import RSIIndicator
from ta.volatility import BollingerBands
from ta.trend import MACD
import seaborn as sns
sns.set(style='whitegrid')

tickers = [
    'ADANIENT.NS', 'ADANIPORTS.NS', 'ASIANPAINT.NS', 'AXISBANK.NS', 'BAJAJ-AUTO.NS',
    'BAJFINANCE.NS', 'BAJAJFINSV.NS', 'BHARTIARTL.NS', 'BPCL.NS', 'BRITANNIA.NS',
    'CIPLA.NS', 'COALINDIA.NS', 'DIVISLAB.NS', 'DRREDDY.NS', 'EICHERMOT.NS',
    'GRASIM.NS', 'HCLTECH.NS', 'HDFCBANK.NS', 'HDFCLIFE.NS', 'HEROMOTOCO.NS',
    'HINDALCO.NS', 'HINDUNILVR.NS', 'ICICIBANK.NS', 'ITC.NS', 'INDUSINDBK.NS',
    'INFY.NS', 'JSWSTEEL.NS', 'KOTAKBANK.NS', 'LT.NS', 'M&M.NS',
    'MARUTI.NS', 'NESTLEIND.NS', 'NTPC.NS', 'ONGC.NS', 'POWERGRID.NS',
    'RELIANCE.NS', 'SBILIFE.NS', 'SBIN.NS', 'SUNPHARMA.NS', 'TATACONSUM.NS',
    'TATAMOTORS.NS', 'TATASTEEL.NS', 'TECHM.NS', 'TITAN.NS', 'ULTRACEMCO.NS',
    'UPL.NS', 'WIPRO.NS', 'TCS.NS', 'APOLLOHOSP.NS', 'HDFC.NS'
]

# Function to calculate technical indicators for each stock
def calculate_indicators(df):
    # Calculate Exponential Moving Averages,
    df['EMA12'] = df['Close'].ewm(span=12, adjust=False).mean()
    df['EMA26'] = df['Close'].ewm(span=26, adjust=False).mean()
    
    # RSI Calculation
    df['RSI'] = RSIIndicator(df['Close']).rsi()
   
    # Bollinger Bands Calculation
    bb = BollingerBands(df['Close'])
    df['BB_High'] = bb.bollinger_hband()
    df['BB_Low'] = bb.bollinger_lband()
    
    # MACD Calculation
    macd = MACD(df['Close'])
    df['MACD'] = macd.macd()
    df['MACD_Signal'] = macd.macd_signal()

    return df

# Function to calculate momentum score for the stock
def calculate_momentum_score(df):
    # RSI Score: Apply non-linear scaling, more emphasis on extreme values
    if df['RSI'].iloc[-1] > 65:
        rsi_score = -1 * ((df['RSI'].iloc[-1] - 50) / 50) ** 2  # Strong negative signal
    elif df['RSI'].iloc[-1] < 35:
        rsi_score = ((df['RSI'].iloc[-1] - 50) / 50) ** 2  # Strong positive signal
    else:
        rsi_score = (df['RSI'].iloc[-1] - 50) / 50  # Linear scaling for normal values

    # MACD Score: Normalize the difference between MACD and Signal line
    macd_diff = df['MACD'].iloc[-1] - df['MACD_Signal'].iloc[-1]
    macd_score = macd_diff / max(abs(df['MACD'].iloc[-1]), abs(df['MACD_Signal'].iloc[-1]), 1e-9)  # Avoid division by zero

    # Bollinger Bands Score: Scale the price relative to Bollinger Bands
    bb_mid = (df['BB_High'].iloc[-1] + df['BB_Low'].iloc[-1]) / 2
    bb_score = (df['Close'].iloc[-1] - bb_mid) / (df['BB_High'].iloc[-1] - df['BB_Low'].iloc[-1])

    # Volatility adjustment: Adjust scores by volatility (ATR, or std)
    volatility = df['Close'].rolling(window=20).std().iloc[-1]  # Standard deviation over 20 days
    volatility_factor = 1 / max(volatility, 1e-9)  # Higher volatility reduces momentum significance

    # Weights for each score - dynamically adjusted
    weights = {
        'RSI': 0.25 * (1 + abs(rsi_score)),  # More weight when RSI is extreme
        'MACD': 0.25 * (1 + abs(macd_score)),  # Increase weight if MACD has a stronger divergence
        'BB': 0.25 * (1 + abs(bb_score))  # Strong divergence from bands increases weight
    }
    
    # Composite Momentum Score: Weighted sum of all individual scores, adjusted for volatility
    composite_score = (
        (weights['RSI'] * rsi_score) +
        (weights['MACD'] * macd_score) +
        (weights['BB'] * bb_score)
    ) * volatility_factor  # Volatility adjustment

    return composite_score

# Dictionary to hold the data for each ticker
data = {}
scores_over_time = {}

# Download data for all tickers and calculate indicators
for ticker in tickers:
    print(f"Fetching data for {ticker}...")
    df = yf.download(ticker, start="2019-01-01", end="2023-12-31")
    df = calculate_indicators(df)  # Calculate technical indicators
    data[ticker] = df  # Store the df in the dictionary

# Time-based analysis: Calculate momentum scores at 5-day intervals
start_date = pd.Timestamp('2019-01-01')
end_date = pd.Timestamp('2023-12-31')

# Start scoring from 20 days after the start date
while start_date < end_date:
    period_end = start_date + pd.Timedelta(days=5)  # Updated to 5 days

    # Calculate scores for this period using the last 20 days up to period_end
    scores = {}
    for ticker, df in data.items():
        # Use the last 20 days of data leading up to period_end for calculations
        period_data = df[(df.index <= period_end)].tail(20)

        # Check if we have enough data points and no NaN values for required indicators
        required_columns = ['RSI', 'MACD', 'MACD_Signal', 'BB_High', 'BB_Low', 'Close']
        if len(period_data) == 20 and not period_data[required_columns].isnull().values.any():
            scores[ticker] = calculate_momentum_score(period_data)

    # If no scores were calculated, assign NaN scores
    if not scores:
        scores = {ticker: np.nan for ticker in tickers}  # Assign NaN
    # Remove all-NaN score entries
    if any(not pd.isna(score) for score in scores.values()):
        # Store the scores for the period only if not all are NaN
        scores_over_time[start_date.strftime('%Y-%m-%d')] = scores

    # Move to the next period (5 days forward)
    start_date = period_end

# Print out the top bullish and bearish stocks for the last analyzed period
latest_scores = list(scores_over_time.values())[-1]
standardized_scores = pd.Series(latest_scores).rank(pct=True)  # Rank from 0 to 1

bullish_stocks_latest = standardized_scores.nlargest(5)
bearish_stocks_latest = standardized_scores.nsmallest(5)

print("\nTop 5 Latest Bullish Stocks  (Standardized Scores):")
for stock, score in bullish_stocks_latest.items():
    print(f"{stock}: Standardized Score = {score:.4f}")

print("\nTop 5 Latest Bearish Stocks (Standardized Scores):")
for stock, score in bearish_stocks_latest.items():
    print(f"{stock}: Standardized Score = {score:.4f}")

# Visualization Function: Plot indicators for one ticker
def plot_stock(ticker):
    df = data[ticker]
    plt.figure(figsize=(14, 8))
    plt.plot(df['Close'], label=f'{ticker} Close Price', color='blue')
    plt.fill_between(df.index, df['BB_Low'], df['BB_High'], color='grey', alpha=0.3, label='Bollinger Bands')
    plt.title(f'{ticker} - Technical Indicators')
    plt.legend()
    plt.show()

# Plot the top bullish stock in the last period
plot_stock(bullish_stocks_latest.index[0])

# Function to generate trading signals based on EMA crossover and RSI
def generate_signals(df):
    df['Trade_Signal'] = 0
    # Buy Signal
    buy_signal = (
        (df['EMA12'] > df['EMA26']) &
        (df['EMA12'].shift(1) <= df['EMA26'].shift(1)) &
        (df['RSI'] > 50)
    )
    # Sell Signal
    sell_signal = (
        (df['EMA12'] < df['EMA26']) &
        (df['EMA12'].shift(1) >= df['EMA26'].shift(1)) &
        (df['RSI'] < 50)
    )
    df.loc[buy_signal, 'Trade_Signal'] = 1  # Buy signal
    df.loc[sell_signal, 'Trade_Signal'] = -1  # Sell signal
    return df

# Apply signal generation to each stock
for ticker in data:
    data[ticker] = generate_signals(data[ticker])
```

```python
# Time-based analysis: Calculate momentum scores at 5-day intervals
start_date = pd.Timestamp('2019-03-02')
end_date = pd.Timestamp('2023-12-31')

# List to hold the signals over time
signals_over_time = []

# Start scoring from 20 days after the start date
while start_date < end_date:
    period_end = min(start_date + pd.Timedelta(days=5), end_date)  # Change the interval to 5 days

    # Calculate scores for this period using the last 20 days up to period_end
    scores = {}
    for ticker, df in data.items():
        # Use the last 20 days of data leading up to period_end for calculations
        period_data = df[df.index <= period_end].tail(20)

        # Check if we have enough data points and no NaN values for required indicators
        required_columns = ['RSI', 'MACD', 'MACD_Signal', 'BB_High', 'BB_Low', 'Close', 'EMA12', 'EMA26']
        if len(period_data) == 20 and not period_data[required_columns].isnull().values.any():
            scores[ticker] = calculate_momentum_score(period_data)

    # Skip the period if no valid scores were calculated
    if not scores:
        start_date = period_end
        continue

    # Standardize scores and rank stocks
    standardized_scores = pd.Series(scores).rank(pct=True)
        
    # Select top 5 bullish and bottom 5 bearish stocks
    bullish_stocks = standardized_scores.nlargest(5)
    bearish_stocks = standardized_scores.nsmallest(5)

    # Ensure a minimum of 3 bearish stocks (if fewer than 5 bearish)
    if len(bearish_stocks) < 5:
        bearish_stocks = standardized_scores.nsmallest(max(3, len(bearish_stocks)))
    
    # Assign weights based on standardized scores
    total_bullish_score = bullish_stocks.sum()
    bullish_weights = (bullish_stocks / total_bullish_score).to_dict()

    total_bearish_score = bearish_stocks.sum()
    bearish_weights = (bearish_stocks / total_bearish_score).to_dict()

    # Generate signals based on adjusted EMA crossover and RSI conditions
    bullish_signals = {}
    for stock in bullish_stocks.index:
        df_stock = data[stock]
        # Find the closest available date to period_end
        if period_end in df_stock.index:
            closest_date = period_end
        else:
            idx = df_stock.index.searchsorted(period_end, side='right') - 1
            if idx >= 0:
                closest_date = df_stock.index[idx]
            else:
                continue  # No valid date found, skip

        # Ensure the required columns are present and not NaN
        if df_stock.loc[closest_date, ['EMA12', 'EMA26', 'RSI']].isnull().any():
            continue  # Skip if any required data is missing

        # Adjusted Buy Signal: Relax the crossover condition and RSI
        ema12 = df_stock.loc[closest_date]['EMA12']
        ema26 = df_stock.loc[closest_date]['EMA26']
        rsi = df_stock.loc[closest_date]['RSI']

        # Buy signal if EMA12 > EMA26 and RSI > 30
        if (ema12 > ema26) and (rsi > 30):
            signal = 1  # Buy signal
        else:
            signal = 0  # No signal

        if signal == 1:
            bullish_signals[stock] = bullish_weights[stock]

    bearish_signals = {}
    for stock in bearish_stocks.index:
        df_stock = data[stock]
        # Find the closest available date to period_end
        if period_end in df_stock.index:
            closest_date = period_end
        else:
            idx = df_stock.index.searchsorted(period_end, side='right') - 1
            if idx >= 0:
                closest_date = df_stock.index[idx]
            else:
                continue  # No valid date found, skip

        # Ensure the required columns are present and not NaN
        if df_stock.loc[closest_date, ['EMA12', 'EMA26', 'RSI']].isnull().any():
            continue  # Skip if any required data is missing

        # Adjusted Sell Signal: Further relax the crossover condition and RSI threshold
        ema12 = df_stock.loc[closest_date]['EMA12']
        ema26 = df_stock.loc[closest_date]['EMA26']
        rsi = df_stock.loc[closest_date]['RSI']

        # Sell signal if EMA12 < EMA26 and RSI < 70
        if (ema12 < ema26) and (rsi < 70):
            signal = -1  # Sell signal
        else:
            signal = 0  # No signal

        if signal == -1:
            bearish_signals[stock] = bearish_weights[stock]

    # Store the signals for the period
    signals_over_time.append({
        'date': period_end,
        'bullish_stocks': bullish_signals,
        'bearish_stocks': bearish_signals
    })

    # Move to the next period
    start_date = period_end

# Print the results for the last period analyzed
print("\nSignals Over Time:")
for signal_info in signals_over_time[-1:]:
    print(f"Date: {signal_info['date']}")
    print("Bullish Stocks with Weights:", signal_info['bullish_stocks'])
    print("Bearish Stocks with Weights:", signal_info['bearish_stocks'])
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from ta.trend import EMAIndicator

# Function to calculate market momentum using Nifty 50 with 20 EMA, 50 EMA, and 200 EMA
def calculate_market_momentum(nifty_data):
    ema_20 = EMAIndicator(close=nifty_data['Close'], window=20).ema_indicator()
    ema_50 = EMAIndicator(close=nifty_data['Close'], window=50).ema_indicator()
    ema_200 = EMAIndicator(close=nifty_data['Close'], window=200).ema_indicator()
    
    # Market is bullish if both EMA 20 > EMA 50 and EMA 50 > EMA 200
    market_momentum = (ema_20.iloc[-1] > ema_50.iloc[-1]) and (ema_50.iloc[-1] > ema_200.iloc[-1])
    return market_momentum

# Backtesting strategy
def backtest_strategy(signals_over_time, data):
    portfolio_value = [1000000]  # Starting with ₹1,000,000
    dates = []
    stop_loss_threshold = 0.1  # 10% stop-loss
    total_long_trades = 0
    total_short_trades = 0
    remaining_trades = {}

    # Download Nifty 50 data for market momentum calculation
    nifty_data = yf.download("^NSEI", start="2019-03-22", end="2023-12-31")

    for signal_info in signals_over_time:
        date = signal_info['date']
        bullish_stocks = signal_info['bullish_stocks']
        bearish_stocks = signal_info['bearish_stocks']
        holdings = {}
        long_cash, short_cash = portfolio_value[-1], portfolio_value[-1]  # Cash allocation

        # Calculate market momentum based on the 20, 50, and 200 EMAs
        market_momentum = calculate_market_momentum(nifty_data[nifty_data.index <= date])

        # Allocate more to long if market is bullish, more to short if bearish
        if market_momentum:  # Market uptrend
            long_allocation = 0.8  # Allocate 80% to long positions
            short_allocation = 0.2  # Allocate 20% to short positions
        else:  # Market downtrend
            long_allocation = 0.2  # Allocate 20% to long positions
            short_allocation = 0.8  # Allocate 80% to short positions

        # Total cash to allocate for long and short positions
        long_cash = long_cash * long_allocation
        short_cash = short_cash * short_allocation

        # Buy bullish stocks (Long positions)
        for stock, weight in bullish_stocks.items():
            df_stock = data[stock]
            if date in df_stock.index:
                price = df_stock.loc[date]['Close']
                allocation = long_cash * weight  # Use the long_cash allocation
                quantity = allocation / price
                holdings[stock] = {'quantity': quantity, 'entry_price': price, 'type': 'long'}
                total_long_trades += 1  # Increment long trade count

        # Short bearish stocks (Short positions)
        for stock, weight in bearish_stocks.items():
            df_stock = data[stock]
            if date in df_stock.index:
                price = df_stock.loc[date]['Close']
                allocation = short_cash * weight  # Use the short_cash allocation
                quantity = -allocation / price  # Negative quantity for short selling
                holdings[stock] = {'quantity': quantity, 'entry_price': price, 'type': 'short'}
                total_short_trades += 1  # Increment short trade count

        # Calculate portfolio return for the next period
        next_date = date + pd.Timedelta(days=10)
        portfolio_return = 0
        for stock, position in holdings.items():
            df_stock = data[stock]
            # Find the exit price at the next_date or the closest available date
            if next_date in df_stock.index:
                exit_date = next_date
            else:
                idx = df_stock.index.searchsorted(next_date, side='right') - 1
                if idx >= 0:
                    exit_date = df_stock.index[idx]
                else:
                    continue  # No exit data available

            exit_price = df_stock.loc[exit_date]['Close']
            entry_price = position['entry_price']

            # Check stop-loss (exit early if price drops more than 10%)
            if (exit_price / entry_price - 1) < -stop_loss_threshold:
                exit_price = entry_price * (1 - stop_loss_threshold)  # Exit at stop-loss price

            # Calculate profit or loss
            if position['type'] == 'long':
                profit = position['quantity'] * (exit_price - entry_price)
            else:  # Short position
                profit = position['quantity'] * (entry_price - exit_price)  # Inverted for shorting

            portfolio_return += profit

        # Update portfolio value
        new_portfolio_value = portfolio_value[-1] + portfolio_return
        portfolio_value.append(new_portfolio_value)
        dates.append(next_date)

        # Store remaining trades for the final square-off at the end
        remaining_trades.update(holdings)

    # Square off remaining trades on the final day (2023-12-31)
    final_date = pd.Timestamp('2023-12-31')
    portfolio_return = 0
    for stock, position in remaining_trades.items():
        df_stock = data[stock]
        if final_date in df_stock.index:
            exit_price = df_stock.loc[final_date]['Close']
        else:
            idx = df_stock.index.searchsorted(final_date, side='right') - 1
            exit_price = df_stock.iloc[idx]['Close'] if idx >= 0 else None

        if exit_price is None:
            continue  # No valid exit price, skip

        entry_price = position['entry_price']

        # Calculate profit or loss for remaining trades
        if position['type'] == 'long':
            profit = position['quantity'] * (exit_price - entry_price)
        else:  # Short position
            profit = position['quantity'] * (entry_price - exit_price)  # Inverted for shorting

        portfolio_return += profit

    # Update portfolio value for the final square-off
    new_portfolio_value = portfolio_value[-1] + portfolio_return
    portfolio_value.append(new_portfolio_value)
    dates.append(final_date)

    # Calculate metrics
    portfolio_returns = pd.Series(np.diff(portfolio_value) / portfolio_value[:-1])

    sharpe_ratio = (portfolio_returns.mean() / portfolio_returns.std()) * np.sqrt(252) if portfolio_returns.std() != 0 else 0
    downside_returns = portfolio_returns[portfolio_returns < 0]
    sortino_ratio = (portfolio_returns.mean() / downside_returns.std()) * np.sqrt(252) if downside_returns.std() != 0 else 0

    running_max = np.maximum.accumulate(portfolio_value)
    drawdown = (portfolio_value - running_max) / running_max
    max_drawdown = drawdown.min()

    nifty_initial_value = nifty_data['Close'].iloc[0]
    nifty_final_value = nifty_data['Close'].iloc[-1]
    nifty_percentage_return = ((nifty_final_value - nifty_initial_value) / nifty_initial_value) * 100

    cumulative_returns = [((value - portfolio_value[0]) / portfolio_value[0]) * 100 for value in portfolio_value]

    # Print results
    final_portfolio_value = portfolio_value[-1]
    percentage_return = ((final_portfolio_value - portfolio_value[0]) / portfolio_value[0]) * 100
    print(f"Final Portfolio Value: ₹{final_portfolio_value:,.2f}")
    print(f"Percentage Return: {percentage_return:.2f}%")
    print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
    print(f"Sortino Ratio: {sortino_ratio:.4f}")
    print(f"Maximum Drawdown: {max_drawdown * 100:.2f}%")
    print(f"Nifty 50 Percentage Return: {nifty_percentage_return:.2f}%")
    print(f"Total Long Trades: {total_long_trades}")
    print(f"Total Short Trades: {total_short_trades}")

    # Fetch Nifty 50 data and normalize it
    nifty_data['Normalized'] = nifty_data['Close'] / nifty_data['Close'].iloc[0] * 100
    nifty_data = nifty_data[nifty_data.index >= dates[0]]  # Filter Nifty data for the same period

    # Normalize portfolio value to start at 100 for comparison
    portfolio_value_normalized = [val / portfolio_value[0] * 100 for val in portfolio_value]

    # Plot portfolio value vs Nifty 50
    plt.figure(figsize=(12, 6))
    plt.plot(dates, portfolio_value_normalized[1:], marker='o', label='Portfolio Value', color='blue')
    plt.plot(nifty_data.index, nifty_data['Normalized'], label='Nifty 50 Index', color='red')
    plt.title('Portfolio Value vs Nifty 50 Index Over Time')
    plt.xlabel('Date')
    plt.ylabel('Normalized Value (Base 100)')
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 6))
    plt.plot(dates, cumulative_returns[1:], marker='o', label='Cumulative Returns', color='green')
    plt.title('Cumulative Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Return (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

```
